```python
import pandas as pd
from collections import OrderedDict

friend_list = [
               ['name', ['John', 'Jenny', 'Nate']],
               ['age',[20,30,30]],
               ['job',['student', 'developer','teacher']]
                
]

df = pd.DataFrame.from_dict(OrderedDict(friend_list))
df
```

```python
## df = df[1:3] 을 하면 df를 변경할 수 있다. 
df.loc[ [0,2] ] ## location, 인덱싱한 행만 출력한다. 
```

```python
## by column condition
##df[df.age>25]  ## 나이가 25보다 많은 사람만 출력. 
##df.query('age>25')  ## 쿼리처럼 쓸 수도 있다. 

##나이가 25이상이고 이름이 네이트인 사람
df[(df.age>25) & (df.name == 'Nate')]
```

```python
friend_list = [
               ['John', 20, 'student'],
               ['Jenny', 30, 'developer'],
               ['Nate', 30 , 'teacher']                
]

df = pd.DataFrame.from_records(friend_list)
df
```

```python
df.iloc[:, 0:2]  ## ilic는 행과 열을 모두 인덱싱한다. 앞이 행(:로 했으므로 전부다), 뒤가 열이다. 
```

```python
df = pd.read_csv('friend_list_no_head.csv', header = None, names = ['name', 'age', 'job'])
df
```

```python
##열의 이름으로 필터링하기

df_filtered = df[['name', 'age']]
df_filtered
```

```python
##filter 함수를 이용
df.filter(items=['age', 'job'])
```

```python
##column 이름에 a가 들어가는 열만 출력. axis=0이면 행, axis=1이면 열에 적용

df.filter(like = 'a', axis=1)
```

```python
df.filter(regex='b$', axis=1) ##b로 끝나는 열만 출력. 만약 n으로 시작하는 것만 출력하려면 ^n
```
