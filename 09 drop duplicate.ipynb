```python
import pandas as pd
```

```python
student_list = [
                {'name': 'John', 'major': 'computer science', 'sex': 'male'},
                {'name': 'Nate', 'major': 'computer science', 'sex': 'male'},
                {'name': 'Edward', 'major': 'computer science', 'sex': 'male'},
                {'name': 'Zara', 'major': 'psychology', 'sex': 'female'},
                {'name': 'John', 'major': 'computer science', 'sex': 'male'}
]

df = pd.DataFrame(student_list, columns = ['name', 'major', 'sex'])
df
```

```python
##중복되는 데이터 확인
df.duplicated()
```

```python
#중복되는 데이터 제거하는 함수
df.drop_duplicates()
```

```python
student_list = [
                {'name': 'John', 'major': 'computer science', 'sex': 'male'},
                {'name': 'Nate', 'major': 'computer science', 'sex': 'male'},
                {'name': 'Edward', 'major': 'computer science', 'sex': 'male'},
                {'name': 'Zara', 'major': 'psychology', 'sex': 'female'},
                {'name': 'John', 'major': 'economics', 'sex': 'male'},
                {'name': 'Nate', 'major': 'economics', 'sex': 'male'}
]

df = pd.DataFrame(student_list, columns = ['name', 'major', 'sex'])
df
```

```python
##이름이 같은 중복 데이터 찾기
df.duplicated(['name'])
```

```python
# 나중에 나오는 동명이인 데이터만 남기기

df.drop_duplicates(['name'], keep = 'last')

#만약 처음에 나오는 데이터를 남기고 싶다면 last가 아니라 first로 하면 된다. keep을 안쓰면 first가 디폴트이다. 
```
