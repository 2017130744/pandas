```python
import pandas as pd
```

```python
date_list = [
             {'date': '2000-06-27'},
             {'date': '2002-09-24'},
             {'date': '2005-12-20'}
]

df = pd.DataFrame(date_list,columns = ['date'])
df
```

```python
def extract_year(date):
  return date.split('-')[0]

df['year'] = df['date'].map(extract_year)
df

##map 함수도 apply와 똑같이 쓸 수 있다. 
```

```python
job_list = [
            {'age': 20, 'job': 'student'},
            {'age': 30, 'job': 'developer'},
            {'age': 20, 'job': 'teacher'}
]

df = pd.DataFrame(job_list)
df
```

```python
df.job = df.job.map({'student':1, "developer": 2, 'teacher':3})
df

##map 함수는 apply함수처럼 column을 변경시킬 수 있고, 딕셔너리를 직접 전달해서 원하는 값으로 변경할 수도 있다. 
```

```python
x_y = [
       {'x':5.5, 'y':-5.6, 'z': -1.1},
       {'x':-5.2, 'y':5.5, 'z': -2.2},
       {'x':-1.6, 'y':-4.5, 'z': -3.3}
]

df = pd.DataFrame(x_y)

df
```

```python
import numpy as np

df = df.applymap(np.around)
## 데이터프레임 전체를 덮는 것, np.around는 반올림
```
