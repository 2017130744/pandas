```python
import pandas as pd
```

```python
a = [{'name': 'John', 'job':'teacher'}, 
      {'name': 'Nate', 'job':'student'},
      {'name': 'Fred', 'job':'developer'}]


b =  [{'name': 'Ed', 'job':'dentist'},
      {'name': 'Jack', 'job':'farmer'},
      {'name': 'Ted', 'job':'designer'}]

c =  [{'name': 'John', 'job':'teacher'},
      {'name': 'Nate', 'job':'student'},
      {'name': 'Jack', 'job':'developer'}]

d = [{'age': 25, 'country':'U.S'},
      {'age': 30, 'country':'U.K'},
      {'age': 45, 'country':'Korea'}]

```

```python
df1 = pd.DataFrame(a, columns = ['name', 'job'])
df2 = pd.DataFrame(b, columns = ['name', 'job'])
```

```python
df1
```

```python
df2
```

```python
#두 테이블 합치기

result = pd.concat([df1, df2])
result

##근데 인덱스가 012012로 된다. 이를 다시 순서대로 맙추려면 ignore_index = True 를 하면 된다. 
```

```python
#또다른 방법
result = df1.append(df2, ignore_index = True)
result
```

```python
#열 합치기
df1 = pd.DataFrame(c, columns = ['name', 'job'])
df2 = pd.DataFrame(d, columns = ['age', 'country'])
```

```python
result = pd.concat([df1, df2], axis = 1, ignore_index = True)
result

## axis만 1로 해주며 ㄴ된다. 
```
