```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = sns.load_dataset('iris')
df
```

```python
df.species.unique()
```

```python
df.max()
```

```python
df1 = df[(df.species == 'setosa')] 
df1.drop('species', axis=1, inplace = True)
df2 = df[(df.species == 'versicolor')]
df2.drop('species', axis=1, inplace = True)
df3 = df[(df.species == 'virginica')]
df3.drop('species', axis=1, inplace = True)

print('setosa\n', df1.max(), '\n')
print('versicolor\n', df2.max(), '\n')
print('virginica\n', df3.max(), '\n')
```

```python
def a(column):
  if column == 'setosa':
    result = 1
  else:
    result = 0

  return result

df['species'] = df['species'].apply(a)

df
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
df = titanic[['pclass', 'age', 'sibsp', 'fare', 'class']]
df

```

```python
column = ['pclass', 'age', 'sibsp', 'fare', 'class']

for i in column:
  print(i, ' : ', df[i].isnull().sum())
```

```python
for i in column:
  if df[i].isnull().sum() !=0:
    df[i].fillna(df[i].mean(), inplace = True)

df.info()
```

```python
survived = pd.read_csv('survived.csv')
survived
```

```python
df2 = pd.concat([df, survived], axis=1)
df2
```

```python
df2.groupby('class')['survived'].value_counts(normalize=True)

## 각 클래스 별 1의 상대적 비율을 보면 된다. 
```
